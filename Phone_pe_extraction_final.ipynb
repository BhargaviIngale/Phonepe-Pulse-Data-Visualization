```python
path1 = r"C:/Users/Bhargavi Ingale/pulse/data/aggregated/transaction/country/india/state"
Agg_state_list=os.listdir(path1)
Agg_state_list

clm={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_state_list:
    p_i= path1+'/'+i
    # print(p_i)
    
    Agg_yr=os.listdir(p_i)
    # print(Agg_yr)
    for j in Agg_yr:
        p_j=p_i+"/"+j
        Agg_yr_list=os.listdir(p_j)
        # print(Agg_yr_list)
        for k in Agg_yr_list:
            p_k=p_j+'/'+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transacion_type'].append(Name)
              clm['Transacion_count'].append(count)
              clm['Transacion_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Trans=pd.DataFrame(clm)
    
    
    
```

```python
path2=r"C:/Users/Bhargavi Ingale/pulse/data/aggregated/user/country/india/state"
Agg_state_list1=os.listdir(path2)
Agg_state_list1

cln={'State':[], 'Year':[],'Quater':[],'Brand':[], 'Count':[], 'Percentage':[]}

for i in Agg_state_list:
    p_i=path2+"/"+i
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+"/"+j
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+'/'+k
            Data=open(p_k,'r')
            B=json.load(Data)
            if 'usersByDevice' in B['data'] and B['data']['usersByDevice'] is not None:
              for z in B['data']['usersByDevice']:
                brand=z['brand']
                count=z['count']
                percentage=z['percentage']
                cln['Brand'].append(brand)
                cln['Count'].append(count)
                cln['Percentage'].append(percentage)
                cln['State'].append(i)
                cln['Year'].append(j)
                cln['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Users=pd.DataFrame(cln)
```

```python
# Top Transactions

path5 =  r"C:/Users/Bhargavi Ingale/pulse/data/top/transaction/country/india/state"
Top_state_list=os.listdir(path5)
Top_state_list

clm4={'State':[], 'Year':[],'Quater':[],'District_Pincode':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Top_state_list:
  p_i=path5+"/"+i
  top_yr=os.listdir(p_i)
  for j in top_yr:
    p_j=p_i+"/"+j
    top_yr_list=os.listdir(p_j)
    for k in top_yr_list:
      p_k=p_j+'/'+k
      data=open(p_k,'r')
      E=json.load(data)
      for z in E['data']['pincodes']:
        name=z['entityName']
        count=z['metric']['count']
        amount=z['metric']['amount']
        clm4['District_Pincode'].append(name)
        clm4['Transaction_count'].append(count)
        clm4['Transaction_amount'].append(amount)
        clm4['State'].append(i)
        clm4['Year'].append(j)
        clm4['Quater'].append(int(k.strip('.json')))

Top_Transaction=pd.DataFrame(clm4)
Top_Transaction
```

```python
# Top users

path6 =r"C:/Users/Bhargavi Ingale/pulse/data/top/user/country/india/state"
Top_state_list1=os.listdir(path6)
Top_state_list1

clm5 = {'State': [], 'Year': [], 'Quater': [], 'District_Pincode': [], 'RegisteredUsers': []}

for i in Top_state_list1:
    p_i = path6  + "/"+i
    top_yr = os.listdir(p_i)
    for j in top_yr:
        p_j = p_i + "/"+j
        top_yr_list = os.listdir(p_j)
        for k in top_yr_list:
            p_k = p_j + '/'+ k
            data = open(p_k, 'r')
            F = json.load(data)
            for z in F['data']['pincodes']:
                name = z['name']
                registeredUsers = z['registeredUsers']
                clm5['District_Pincode'].append(name)
                clm5['RegisteredUsers'].append(registeredUsers)
                clm5['State'].append(i)
                clm5['Year'].append(j)
                clm5['Quater'].append(int(k.strip('.json')))

Top_Users=pd.DataFrame(clm5)
Top_Users
```
